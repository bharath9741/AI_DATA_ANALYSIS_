```python
# Part 1: Handle Missing Values & Duplicates

    # Step-by-Step Guidelines:
# 1. Load the Data: First, ensure you have pandas installed and import it.
# 2. Handling Missing Values
#     1. Identify Missing Values:
#     2. Fill Missing Values:
# 3. Handling Duplicates
#     1. Identify Duplicates:
#     2. Remove Duplicates:
# 4. Combined Practice on a New Dataset
#     1. New Sample Data:
#     2. Handling Missing Values:
#     3. Remove Duplicates:

import pandas as pd
import numpy as np

# 1. Load the Data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', None],
    'Age': [25, None, 30, 25, None, 22],
    'City': ['NY', 'LA', 'NY', 'NY', 'LA', 'LA']
}
df = pd.DataFrame(data)

# 2. Handling Missing Values
# 2.1 Identify Missing Values
print("Missing Values:\n", df.isnull().sum())

# 2.2 Fill Missing Values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Name'] = df['Name'].fillna('Unknown')

# 3. Handling Duplicates
# 3.1 Identify Duplicates
print("Duplicates:\n", df.duplicated())

# 3.2 Remove Duplicates
df = df.drop_duplicates()

# 4. Combined Practice on New Dataset
new_data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', None],
    'Price': [1000, 500, 300, 1000, 200]
}
df2 = pd.DataFrame(new_data)

# 4.2 Handling Missing Values
df2['Product'] = df2['Product'].fillna('Unknown')

# 4.3 Remove Duplicates
df2 = df2.drop_duplicates()

        
        
        
        
        
        
        
        
        
```
