```python
import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Create a bar plot
plt.bar(products, sales, color='skyblue')

# Add title and labels
plt.title('Sales of Products')
plt.xlabel('Product')
plt.ylabel('Sales (in thousands)')

# Display the plot
plt.show()


```
