```python
# Data Profiling Techniques Examples

# 1. Descriptive Statistics:
# Task 1: Calculate the mean, median, and mode for sales figures in a retail dataset.
# Task 2: Analyze the average age, median, and mode in a customer demographic
# dataset.
# Task 3: Determine the mean, median, and mode of daily website visit counts.






# 2. Distribution Analysis:
# Task 1: Visualize the distribution of product prices using histograms.
# Task 2: Create a distribution chart for exam scores in an educational dataset.
# Task 3: Analyze the distribution of order quantities for inventory management.






# 3. Data Type Validation:
# Task 1: Validate numeric fields in a dataset to ensure they contain only numbers.
# Task 2: Check for valid date formats in a transaction log.
# Task 3: Validate email formats in a customer contact dataset.


import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

# =======================
# Task 1: Descriptive Statistics
# =======================

# Task 1.1: Calculate the mean, median, and mode for sales figures in a retail dataset.
def descriptive_statistics_task1():
    sales_data = {'transaction_id': [101, 102, 103, 104],
                  'sales': [250, 300, 350, 400]}
    df_sales = pd.DataFrame(sales_data)
    
    mean_sales = df_sales['sales'].mean()
    median_sales = df_sales['sales'].median()
    mode_sales = df_sales['sales'].mode()[0]
    
    print(f"Sales Descriptive Statistics:")
    print(f"Mean Sales: {mean_sales}")
    print(f"Median Sales: {median_sales}")
    print(f"Mode Sales: {mode_sales}")

# Task 1.2: Analyze the average age, median, and mode in a customer demographic dataset.
def descriptive_statistics_task2():
    customer_data = {'customer_id': [1, 2, 3, 4],
                     'age': [25, 30, 35, 40]}
    df_customers = pd.DataFrame(customer_data)
    
    mean_age = df_customers['age'].mean()
    median_age = df_customers['age'].median()
    mode_age = df_customers['age'].mode()[0]
    
    print(f"\nCustomer Demographic Descriptive Statistics:")
    print(f"Mean Age: {mean_age}")
    print(f"Median Age: {median_age}")
    print(f"Mode Age: {mode_age}")

# Task 1.3: Determine the mean, median, and mode of daily website visit counts.
def descriptive_statistics_task3():
    website_data = {'date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
                    'visit_count': [1000, 1500, 1200, 1300]}
    df_website = pd.DataFrame(website_data)
    
    mean_visits = df_website['visit_count'].mean()
    median_visits = df_website['visit_count'].median()
    mode_visits = df_website['visit_count'].mode()[0]
    
    print(f"\nWebsite Visit Count Descriptive Statistics:")
    print(f"Mean Visits: {mean_visits}")
    print(f"Median Visits: {median_visits}")
    print(f"Mode Visits: {mode_visits}")

# =======================
# Task 2: Distribution Analysis
# =======================

# Task 2.1: Visualize the distribution of product prices using histograms.
def distribution_analysis_task1():
    product_data = {'product_id': [1, 2, 3, 4, 5],
                    'price': [50, 100, 150, 200, 250]}
    df_products = pd.DataFrame(product_data)
    
    plt.hist(df_products['price'], bins=5, edgecolor='black')
    plt.title('Product Price Distribution')
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.show()

# Task 2.2: Create a distribution chart for exam scores in an educational dataset.
def distribution_analysis_task2():
    exam_data = {'student_id': [1, 2, 3, 4, 5],
                 'score': [90, 80, 70, 85, 95]}
    df_scores = pd.DataFrame(exam_data)
    
    plt.hist(df_scores['score'], bins=5, edgecolor='black')
    plt.title('Exam Score Distribution')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.show()

# Task 2.3: Analyze the distribution of order quantities for inventory management.
def distribution_analysis_task3():
    inventory_data = {'product_id': [1, 2, 3, 4, 5],
                      'order_quantity': [100, 150, 200, 250, 300]}
    df_inventory = pd.DataFrame(inventory_data)
    
    plt.hist(df_inventory['order_quantity'], bins=5, edgecolor='black')
    plt.title('Order Quantity Distribution')
    plt.xlabel('Order Quantity')
    plt.ylabel('Frequency')
    plt.show()

# =======================
# Task 3: Data Type Validation
# =======================

# Task 3.1: Validate numeric fields in a dataset to ensure they contain only numbers.
def data_type_validation_task1():
    data = {'product_id': [1, 2, 'three', 4, 5], 'price': [100, 200, 300, 'four hundred', 500]}
    df = pd.DataFrame(data)
    
    numeric_check = df.apply(pd.to_numeric, errors='coerce')  # Convert non-numeric to NaN
    invalid_numeric_rows = numeric_check.isna().sum()
    
    print(f"\nInvalid Numeric Data (Rows with NaN):\n{invalid_numeric_rows}")

# Task 3.2: Check for valid date formats in a transaction log.
def data_type_validation_task2():
    transaction_data = {'transaction_id': [101, 102, 103],
                        'transaction_date': ['2023-01-01', '2023-02-30', '2023-03-15']}
    df_transactions = pd.DataFrame(transaction_data)
    
    df_transactions['transaction_date'] = pd.to_datetime(df_transactions['transaction_date'], errors='coerce')
    invalid_dates = df_transactions[df_transactions['transaction_date'].isna()]
    
    print(f"\nInvalid Dates in Transaction Log:\n{invalid_dates}")

# Task 3.3: Validate email formats in a customer contact dataset.
def data_type_validation_task3():
    contact_data = {'customer_id': [1, 2, 3],
                    'email': ['customer1@example.com', 'customer2example.com', 'customer3@domain.com']}
    df_contacts = pd.DataFrame(contact_data)
    
    email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'  # Regular expression for email validation
    df_contacts['valid_email'] = df_contacts['email'].apply(lambda x: bool(re.match(email_pattern, x)))
    invalid_emails = df_contacts[~df_contacts['valid_email']]
    
    print(f"\nInvalid Emails:\n{invalid_emails}")

# =======================
# Run All Tasks
# =======================
def run_all_tasks():
    print("Running Descriptive Statistics Tasks...")
    descriptive_statistics_task1()
    descriptive_statistics_task2()
    descriptive_statistics_task3()
    
    print("\nRunning Distribution Analysis Tasks...")
    distribution_analysis_task1()
    distribution_analysis_task2()
    distribution_analysis_task3()
    
    print("\nRunning Data Type Validation Tasks...")
    data_type_validation_task1()
    data_type_validation_task2()
    data_type_validation_task3()

run_all_tasks()





```
