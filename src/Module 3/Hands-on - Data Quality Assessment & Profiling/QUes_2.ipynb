```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Titanic Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the 'Age' column.







# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Rows in Titanic Dataset
# - Identify any duplicate rows in the dataset.










# Part 3: Generate a Data Quality Report

# Task 3: Titanic Dataset Overview
# - Create a simple report of missing values, duplicates, and some basic statistics for the Titanic dataset.



import pandas as pd

# Part 1: Load the Titanic Dataset & Check Missing Values
# Load the Titanic dataset (replace with your local file path or use a URL)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Check for missing values in the 'Age' column
missing_age = df['Age'].isnull().sum()

# Display the result for missing values in 'Age' column
print(f"Number of missing values in 'Age' column: {missing_age}")

# Part 2: Identify Duplicates & Inconsistencies
# Identify duplicate rows in the Titanic dataset
duplicates = df[df.duplicated()]

# Display duplicate rows (if any)
print("\nDuplicate rows in the Titanic dataset:")
print(duplicates)

# Part 3: Generate a Data Quality Report
# Check for missing values across all columns
missing_values = df.isnull().sum()

# Identify number of duplicate rows
duplicate_rows = df.duplicated().sum()

# Basic statistics for numerical columns
basic_stats = df.describe()

# Generate and display the data quality report
print("\nData Quality Report:")
print(f"Total number of records: {len(df)}")
print(f"Number of duplicate rows: {duplicate_rows}")
print("\nMissing values in each column:")
print(missing_values)
print("\nBasic statistics for numerical columns:")
print(basic_stats)


```
