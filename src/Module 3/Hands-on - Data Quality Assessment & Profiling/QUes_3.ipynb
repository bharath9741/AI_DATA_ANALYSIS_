```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.








# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.









# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.



import pandas as pd

# ---------------------------------------------
# Part 1: Load a Dataset & Check Missing Values
# ---------------------------------------------

# Load the Iris dataset from a URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Check for missing values in the entire dataset
missing_values = df.isnull().sum()

print("=== Missing Values in Iris Dataset ===")
print(missing_values)

# ------------------------------------------------------
# Part 2: Identify Duplicates & Inconsistencies (Example)
# ------------------------------------------------------

# Create a sample DataFrame with inconsistent 'Gender' values
sample_data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['M', 'Male', 'male', 'F', 'Female']
})

# Standardize the 'Gender' column to lowercase and strip whitespace
sample_data['Gender_cleaned'] = sample_data['Gender'].str.strip().str.lower()

# Show unique values in the cleaned 'Gender' column
inconsistent_genders = sample_data['Gender_cleaned'].unique()

print("\n=== Inconsistent Gender Entries (Cleaned) ===")
print(inconsistent_genders)

# ---------------------------------------------
# Part 3: Generate a Data Quality Report (Iris)
# ---------------------------------------------

# Total number of records
total_records = len(df)

# Number of duplicate rows
duplicate_rows = df.duplicated().sum()

# Basic statistics
basic_stats = df.describe()

# Display the report
print("\n=== Data Quality Report: Iris Dataset ===")
print(f"Total records: {total_records}")
print(f"Number of duplicate rows: {duplicate_rows}")
print("\nBasic Descriptive Statistics:")
print(basic_stats)

```
